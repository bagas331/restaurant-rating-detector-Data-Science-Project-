```python
x = df[["Average Cost for two", "Has Table booking", "Has Online delivery", "Price range"]]
```

```python
# ABOVE 2 BELOW 2.5  POOR
# ABOVE 2.5 BELOW 3.5 AVERAGE
# ABOVE 3.5 BELOW 4.0 GOOD
# ABOVE 4 BELOW 4.5 VERY GOOD
# ABOVE 4.5 EXCELLENT
```

```python
y = df[["Aggregate rating"]]
```

```python
x
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
x["Has Table booking"] = labelencoder.fit_transform(x["Has Table booking"])
```

```python
x
```

```python
x["Has Online delivery"] = labelencoder.fit_transform(x["Has Online delivery"])
```

```python
x
```

```python
# HAS TABLE BOOKING 0 IS NO, 1 IS YES
# HAS ONLINE DELIVERY 0 IS NO, 1 IS YES
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
def modelresult(preductions):
  print("Mean absolute error on model is{}".format (mean_absolute_error(y_test, preductions)))
  print("Mean squared error on model is{}".format (mean_squared_error(y_test, preductions)))
  print("Root mean squared error on model is{}".format (np.sqrt(mean_squared_error(y_test, preductions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predictionfromlr = lr.predict(x_test)
```

```python
modelresult(predictionfromlr)
```

```python
predictionfromlr
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {
    "C": [0.1, 0.5],
    "kernel": ["linear", "poly"],
    "degree": [2]
}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(x_train,y_train)
```

```python
predsgridsvr = gridsvr.predict(x_test)
```

```python
modelresult(predsgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = treemodel, param_grid = param_grid)
```

```python
grid_tree.fit(x_train, y_train)
```

```python
treepredictions = grid_tree.predict(x_test)
```

```python
modelresult(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfmodel = RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth" : [5,15],
    "n_estimators" : [2,5,10]
}
```

```python
gridrfr = GridSearchCV(rfmodel, param_gridrfr)
```

```python
gridrfr.fit(x_train, y_train)
```

```python
randomforestpredictions = gridrfr.predict(x_test)
```

```python
modelresult(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
grid_tree.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors": [9,11,13,15]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```

```python
knn_grid_search.fit(x_train, y_train)
```

```python
knnpreds = knn_grid_search.predict(x_test)
```

```python
modelresult(knnpreds)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada_param_grid = {"n_estimators": [50,100,150],
                  "learning_rate": [0.1,0.5,1]}
ada_grid_search = GridSearchCV(ada, ada_param_grid)
```

```python
ada_grid_search.fit(x_train, y_train)
```

```python
adapreds = ada_grid_search.predict(x_test)
```

```python
modelresult(adapreds)
```
